```python
def clean_ingredients(x):
    # 1. If it‚Äôs already a list, keep it
    if isinstance(x, list):
        return x
    # 2. If it‚Äôs missing or a float/NaN, convert to empty list
    if pd.isna(x) or isinstance(x, float):
        return []
    # 3. If it‚Äôs a string that looks like a list, parse it
    if isinstance(x, str):
        try:
            # literal_eval will turn "['a','b']" into ['a','b']
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            # fallback: split on commas
            return [item.strip() for item in x.strip("[]").split(",") if item]
    # 4. Otherwise, cast to string in a one-element list
    return [str(x)]

# 1. Clean the column
df['list_ingredients_T-1'] = df['list_ingredients_T-1'].apply(clean_ingredients)
df['list_ingredients_T-2'] = df['list_ingredients_T-2'].apply(clean_ingredients)
```

```python
from mlxtend.preprocessing import TransactionEncoder

def OneHot(dataset) -> pd.DataFrame:

    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df_exploded = pd.DataFrame(te_ary, columns=te.columns_)

    return df_exploded
```

```python
from mlxtend.frequent_patterns import apriori

MIN_SUPPORT = 0.01

def GetFreqItemset(df: pd.DataFrame) -> pd.DataFrame:

    frequent_itemsets = apriori(df, min_support=MIN_SUPPORT, use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

    return frequent_itemsets.sort_values(by='support', ascending=False)

def freq_set(df: pd.DataFrame) -> pd.DataFrame:

    df_exploded = OneHot(df)
    
    return GetFreqItemset(df_exploded)

```

```python
from mlxtend.frequent_patterns import association_rules

rule0 = association_rules(freq_set(ingre_df0), min_threshold=0.2, metric="confidence")
rule1 = association_rules(freq_set(ingre_df1), min_threshold=0.2, metric="confidence")
rule2 = association_rules(freq_set(ingre_df2), min_threshold=0.2, metric="confidence")
```

```python
def clean_and_filter_rules(rules_df, exclude_items=['‡∏Ç‡πâ‡∏≤‡∏ß']):
    """Clean frozensets and filter out rules containing specified items"""
    
    # Filter out rows where antecedents or consequents contain excluded items
    mask = True
    for item in exclude_items:
        mask &= ~rules_df['antecedents'].apply(lambda x: item in x)
        mask &= ~rules_df['consequents'].apply(lambda x: item in x)
    
    filtered_df = rules_df[mask].copy()
    
    # Convert frozensets to comma-separated strings
    filtered_df['antecedents_list'] = filtered_df['antecedents'].apply(
        lambda x: ', '.join(sorted(list(x)))
    )
    
    filtered_df['consequents_list'] = filtered_df['consequents'].apply(
        lambda x: ', '.join(sorted(list(x)))
    )
    
    # Create clean export DataFrame
    export_df = filtered_df[['antecedents_list', 'consequents_list', 'support', 'confidence', 'lift']].copy()
    export_df.columns = ['Antecedents', 'Consequents', 'Support', 'Confidence', 'Lift']
    
    return export_df

clean_and_filter_rules(res_m1)
```
