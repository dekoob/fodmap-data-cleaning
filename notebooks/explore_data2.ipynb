```python
import ast

ingre_df0 = df_filtered['list_ingredients'].apply(ast.literal_eval)

ingre_df1 = df_filtered['list_ingredients_T-1'].dropna().apply(ast.literal_eval)

ingre_df2 = df_filtered['list_ingredients_T-2'].dropna().apply(ast.literal_eval)
```

```python
from mlxtend.preprocessing import TransactionEncoder

def OneHot(dataset) -> pd.DataFrame:

    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df_exploded = pd.DataFrame(te_ary, columns=te.columns_)

    return df_exploded
```

```python
from mlxtend.frequent_patterns import apriori

MIN_SUPPORT = 0.02

def GetFreqItemset(df: pd.DataFrame) -> pd.DataFrame:

    frequent_itemsets = apriori(df, min_support=MIN_SUPPORT, use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

    return frequent_itemsets.sort_values(by='support', ascending=False)

def freq_set(df: pd.DataFrame) -> pd.DataFrame:

    df_exploded = OneHot(df)
    
    return GetFreqItemset(df_exploded)

```

Find apriori

```python
ingre_df0 = ingre_df0.apply(lambda x: x + ['bloating'])
ingre_df1 = ingre_df1.apply(lambda x: x + ['bloating'])
ingre_df2 = ingre_df2.apply(lambda x: x + ['bloating'])
```

```python
from mlxtend.frequent_patterns import association_rules

rule0 = association_rules(freq_set(ingre_df0), metric="confidence")
rule1 = association_rules(freq_set(ingre_df1), metric="confidence")
rule2 = association_rules(freq_set(ingre_df2), metric="confidence")
```
